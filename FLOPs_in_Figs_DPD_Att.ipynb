```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file
df = pd.read_csv('output.csv')

# Convert FLOPs column to a common unit (e.g., FLOPs in gigabytes)
def convert_flops_to_gb(value):
    value = float(value[:-1])  # Extract the numerical value, excluding the last character 'G'
    return value

# Apply the conversion function to the FLOPs column
df['FLOPs_GB'] = df['FLOPs'].apply(convert_flops_to_gb)

# Plot the variation of the FLOPs column
plt.figure(figsize=(10, 6))
plt.plot(df['FLOPs_GB'], marker='o', linestyle='-', color='b')  # Remove the x values (df['filename'])
plt.ylabel('FLOPs (GB)')
plt.title('Variation of FLOPs')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file
df = pd.read_csv('output.csv')

# Convert FLOPs column to a common unit (e.g., FLOPs in gigabytes)
def convert_flops_to_gb(value):
    value = float(value[:-1])  # Extract the numerical value, excluding the last character 'G'
    return value

# Apply the conversion function to the FLOPs column
df['FLOPs_GB'] = df['FLOPs'].apply(convert_flops_to_gb)

# Plot a scatter plot for the variation of the FLOPs column
plt.figure(figsize=(10, 6))
plt.scatter(df.index, df['FLOPs_GB'], color='red', marker='o')
plt.xlabel('Index')
plt.ylabel('FLOPs (GB)')
plt.title('Variation of FLOPs')
plt.show()
```
